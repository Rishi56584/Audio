```python
# === Cell 1 ===
from moviepy.editor import VideoFileClip
from pydub import AudioSegment
import cv2
import os
import random
import urllib.request
import shutil
import gradio as gr
import moviepy.editor as mp
from deepface import DeepFace

# --- Extract audio from video ---
def extract_audio(video_path):
    video = VideoFileClip(video_path)
    audio_path = "/content/extracted_audio.wav"
    # Extract audio from the video
    os.system(f"ffmpeg -i {video_path} -q:a 0 -map a {audio_path}")

    print(f"Audio extracted successfully: {audio_path}")
    video.audio.write_audiofile(audio_path)
    return audio_path

# --- Simulate vocal separation (placeholder) ---
from spleeter.separator import Separator

import time
import glob

def separate_audio(audio_path):
    output_dir = "/content/separated_audio"
    from spleeter.separator import Separator
    import os

    # Define paths
    #audio_path = "/content/audio.wav"
    #output_dir = "/content/separated_audio"

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Initialize Spleeter with a 2-stem model (voice and background separation)
    separator = Separator('spleeter:2stems')

    # Process the audio file
    separator.separate_to_file(audio_path, output_dir)

    print(f"Audio separation completed. Files saved in {output_dir}")
    vv='/content/separated_audio/extracted_audio/vocals.wav'
    return vv




# --- Extract frames from video every 5 seconds ---
def extract_frames(video_path, interval_sec=15):
    frames = []
    output_dir = "/content/frames"
    os.makedirs(output_dir, exist_ok=True)

    vidcap = cv2.VideoCapture(video_path)
    fps = int(vidcap.get(cv2.CAP_PROP_FPS))  # Get frames per second
    frame_interval = fps * interval_sec       # Calculate frame interval for 5 seconds

    success, image = vidcap.read()
    count = 0
    frame_index = 0
    while success:
        if count % frame_interval == 0:
            frame_path = os.path.join(output_dir, f"frame{frame_index}.jpg")
            cv2.imwrite(frame_path, image)
            frames.append(frame_path)
            frame_index += 1
        success, image = vidcap.read()
        count += 1

    vidcap.release()
    return frames

# --- Detect emotion using DeepFace ---
def detect_emotion_from_frames(frames_folder):
    emotion_counts = {}
    for frame in sorted(os.listdir(frames_folder)):
        frame_path = os.path.join(frames_folder, frame)
        try:
            analysis = DeepFace.analyze(frame_path, actions=['emotion'], enforce_detection=False)
            dominant_emotion = analysis[0]['dominant_emotion']
            emotion_counts[dominant_emotion] = emotion_counts.get(dominant_emotion, 0) + 1
            print(f"{frame} ➝ {dominant_emotion}")
        except:
            print(f"{frame} ➝ No face detected")
    # Return the most frequent detected emotion
    if emotion_counts:
        return max(emotion_counts, key=emotion_counts.get)
    else:
        return "neutral"

# --- Emotion to music mapping ---
def emotion_to_music(emotion):
    if emotion in ["happy", "surprise"]:
        return "upbeat_music.mp3"
    elif emotion == "sad":
        return "melancholic_music.mp3"
    elif emotion == "angry":
        return "intense_music.mp3"
    elif emotion == "fear":
        return "suspenseful_music.mp3"
    elif emotion in ["neutral", "calm"]:
        return "calm_music.mp3"
    else:
        return "default_music.mp3"

# --- Download background music files ---
music_files = {
    "upbeat_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-1.mp3",
    "slow_emotional_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-2.mp3",
    "intense_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-3.mp3",
    "suspenseful_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-4.mp3",
    "default_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-5.mp3",
    "calm_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-6.mp3",
    "melancholic_music.mp3": "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-7.mp3"
}

for filename, url in music_files.items():
    urllib.request.urlretrieve(url, f"/content/{filename}")
```

```python
# === Cell 2 ===

# --- Full video processing pipeline ---
def run_full_pipeline(video_path):
    # Step 1: Extract & separate audio
    audio_path = extract_audio(video_path)
    vocals_path = separate_audio(audio_path)

    # Step 2: Extract frames and detect emotion
    frames = extract_frames(video_path)
    frames_folder = "/content/frames"
    emotion = detect_emotion_from_frames(frames_folder)
    print(f"✅ Detected Emotion: {emotion}")

    # Step 3: Load video to get duration
    video = mp.VideoFileClip(video_path)
    video_duration_ms = int(video.duration * 1000)

    # Step 4: Select background music
    music_file = emotion_to_music(emotion)
    print(f"🎵 Selected Music: {music_file}")
    background = AudioSegment.from_file(f"/content/{music_file}")
    background = background[:video_duration_ms]

    # Step 5: Load vocals, adjust and combine
    vocals = AudioSegment.from_file(vocals_path)
    vocals = vocals[:video_duration_ms]
    background_lowered = background - 5
    combined = background_lowered.overlay(vocals)
    combined = combined[:video_duration_ms]

    # Step 6: Export final audio
    combined_audio_path = "/content/final_combined_audio.mp3"
    combined.export(combined_audio_path, format="mp3")
    print("✅ Final audio exported.")

    # Step 7: Merge final audio with video
    output_video_path = "/content/final_video_with_music.mp4"
    final_audio = mp.AudioFileClip(combined_audio_path)
    final_video = video.set_audio(final_audio)
    final_video.write_videofile(output_video_path, codec="libx264", audio_codec="aac")
    return output_video_path



# --- Gradio Integration ---
def process_video(video_file):
    input_path = "/content/input_video.mp4"
    shutil.copy(video_file, input_path)
    return run_full_pipeline(input_path)

interface = gr.Interface(
    fn=process_video,
    inputs=gr.Video(label="Upload a Video"),
    outputs=gr.Video(label="Final Video with Emotion-Based Music"),
    title="AI-Based Music Composition Using Transformer Network",
    description="Upload a video, The system detects facial emotions and overlays emotion-matching background music."
)

interface.launch(share=True, debug=True)
```
